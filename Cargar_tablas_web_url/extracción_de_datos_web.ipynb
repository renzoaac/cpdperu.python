```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

# URL de la página de Wikipedia
url = "https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_por_PIB_(nominal)"

# Obtener el contenido HTML de la página
response = requests.get(url)
response.encoding = 'utf-8'  # Asegurar la correcta codificación en español
soup = BeautifulSoup(response.text, 'html.parser')

# Extraer todas las tablas de la página
tables = pd.read_html(str(soup), encoding='utf-8')

# Guardar cada tabla en un archivo CSV
for i, table in enumerate(tables):
    table.to_csv(f'tabla_{i}.csv', index=False, encoding='utf-8')
    print(f'Tabla {i} guardada como tabla_{i}.csv')
```
